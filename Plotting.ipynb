```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df
```

```python
plt.scatter(df['sepal_length'],df['species'],color = 'pink')
```

```python
sns.lineplot(x='sepal_length',y='petal_length',data = df, color = 'orange')
```

```python
sns.distplot(df['sepal_width'],color = 'yellow')
```

```python
sns.barplot(x='petal_width',y='sepal_width',data=df)
```

```python
sns.barplot(x='sepal_length',y='petal_length',data=df,color='blue')
```

```python
sns.barplot(x='petal_length',y='species',data=df,color='orange')
```

```python
sns.boxplot(x='petal_length',y='species',data=df,color='black')
```

```python
df.describe()
```

```python
sns.catplot(x='petal_length',y='sepal_length',hue='species',data=df)
```

```python
sns.pairplot(df,hue='species')
```
